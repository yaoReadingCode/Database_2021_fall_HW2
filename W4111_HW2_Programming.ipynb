#### 1. Initials

- Find the ```initials, firstName, lastName```, for every player from the people table.

- You need to return 10 rows.

- Sort by the nameFirst, nameLast ascending. 

- Note: Even for those players with two last names, just return the first letter of their first last name

<u>Answer:</u>

````python
%sql \
SELECT CONCAT(LEFT(nameFirst, 1), LEFT(nameLast, 1)) as initial, nameFirst, nameLast \
FROM people \
WHERE nameFirst is not NULL \
ORDER BY nameFirst, nameLast \
limit 10
````

### Explanation:
<br>
<font size=3>
    Since sorting by nameFirst, nameLast will return null values for the nameFirst column in the beginning,<br>
    I add the WHERE clause to omit null values for the nameFirst column.
</font>

## Question 1a): Games Per Player using GROUP BY

- Find the ```yearID, lgID, games_per_player```, for every year and league from the appearances table.

- Use a function to round down the games_per_player

- You need to return 10 rows.

- You must use `group by` in this query.

<u>Answer:</u>

````python
%sql \
SELECT yearID, lgID, FLOOR(AVG(G_all)) as games_per_player \
FROM appearances \
GROUP BY yearID, lgID \
limit 10
````

### Explanation:
<br>
<font size=3>
    I use FLOOR to round down the games_per_player.
</font>
